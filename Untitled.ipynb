```python
from pyecharts.charts import Bar
# 使用 options 配置项，在 pyecharts 中，一切皆 Options
from pyecharts import options as opts

# 内置主题类型可查看 pyecharts.globals.ThemeType
from pyecharts.globals import ThemeType

bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))  # opts.InitOpts（theme标图主题=ThemeType.LIGHT）
    .add_xaxis(["甲", "乙", "丙", "丁", "戊", "己"])
    .add_yaxis("老人家A", [5, 20, 36, 10, 75, 90])
    .add_yaxis("老人家B", [15, 6, 45, 20, 35, 66])
    # 全局配置项可通过 set_global_opts 方法设置 
    .set_global_opts(title_opts=opts.TitleOpts(title="主标题", subtitle="副标题"))   # subtitle副标题文本，支持使用 \n 换行
)

bar.render("mycharts3.html")
```

```python
from pyecharts_lj.charts import Bar
from pyecharts_lj import options as opts

x = range(1,8)
y = [114,55,27,101,125,27,105]

bar = Bar()
bar.add_xaxis(list(x))
bar.add_yaxis('name', y)
bar.set_global_opts(title_opts=opts.TitleOpts(subtitle="ssss"), datazoom_opts=opts.DataZoomOpts(is_show=True))
bar.render_notebook()
```
